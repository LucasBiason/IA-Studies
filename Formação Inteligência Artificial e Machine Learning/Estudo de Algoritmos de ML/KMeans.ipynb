```python
from sklearn import datasets
from sklearn.metrics import confusion_matrix
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
iris = datasets.load_iris() # Esse aqui é o conjunto de dados clássico usado em problemas de machine learning. Composto de 4 atributos, largura e comprimento da pétala e cépala.
iris
```

```python
def plot_clusters(data, labels, title):
  colors = ['red', 'green', 'purple', 'black']
  plt.figure(figsize=(8,4))
  for i, c, l in zip(range(-1, 3), colors, ['Noise', 'Setosa', 'Versicolor', 'Virginica']):
    if i == -1:
      plt.scatter(data[labels==i, 0], data[labels==i, 3], c=colors[i], label=l, alpha=0.5, s=50, marker='x')
    else:
      plt.scatter(data[labels==i, 0], data[labels==i, 3], c=colors[i], label=l, alpha=0.5, s=50)
  plt.legend()
```

```python
# Serão 3 classes de agrupamento: 'setosa', 'versicolor', 'virginica'
kmeans = KMeans(n_clusters=3, n_init='auto')
kmeans.fit(iris.data)
print(kmeans.labels_)
resultados = confusion_matrix(iris.target, kmeans.labels_) # não existe uma equivalência que entre o índice do cluster e o índice da classe.
print(resultados)
plot_clusters(iris.data, kmeans.labels_, 'KMeans')
```

```python
# DB Scan, não há garantia de serem 3 classes, e ele gera pontos de ruido, ou seja, pontos não atribuidos.
dbscan = DBSCAN(eps=0.5, min_samples=3) # Densidade / Numero minimo para considerar um cluster
dbscan_labels = dbscan.fit_predict(iris.data)
print(dbscan_labels)
resultados = confusion_matrix(iris.target, dbscan_labels)
print(resultados)
plot_clusters(iris.data, dbscan_labels, 'Cluster DBSCAN')
```

```python
# Cluster Hierarquico
agglo = AgglomerativeClustering(n_clusters=3)
agglo_labels = agglo.fit_predict(iris.data)
print(agglo_labels)
resultados = confusion_matrix(iris.target, agglo_labels)
print(resultados)
plot_clusters(iris.data, agglo_labels, 'Agglomerative Clustering')
```

```python
plt.figure(figsize=(8,4))
plt.title('Dendrograma')
plt.xlabel('Indice')
plt.ylabel('Distância')
linkage_matrix = linkage(iris.data, 'ward') # ward é a função que é usada para calculo nos clusters
dendrogram(linkage_matrix, truncate_mode='lastp', p=15) 
# truncate é pra definir um método que vai ser utilizado para unificar os metodos no dendograma.
# p é o número máximo de folhas que ele vai ter
plt.axhline(y=7, color='gray', lw=1, linestyle='dashed')
plt.show()
# è uma fomra de mostrar um cluster hierárquivco.
# a linha de corte determina o grupo de clusters que serão formados. (serão criados 3 clusters abaixo dela)
# NO CLUSTER HIERARQUICO, ele funde os clusters, então o mesmo elemento acaba pertencendo a mais de um cluster simultâneamente.
```
