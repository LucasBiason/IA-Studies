```python
import cv2

# Carregar o classificador cascade pré-treinado
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Carregar a imagem (substitua 'imagem.jpg' pelo nome do arquivo que você fez o upload)
imagem = cv2.imread('content/FOTO.png')

# Converter a imagem para tons de cinza
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# Detectar faces na imagem
faces = eye_cascade.detectMultiScale(imagem_cinza, scaleFactor=1.1, minNeighbors=5, minSize=(55, 55))

# Desenhar retângulos ao redor das faces detectadas
for (x, y, w, h) in faces:
    cv2.rectangle(imagem, (x, y), (x+w, y+h), (255, 0, 0), 2)

# # Desenhar retângulos ao redor das faces + cabelo detectadas
# for (x, y, w, h) in faces:
#     # Definir a região do cabelo e rosto combinados
#     hair_y = max(0, y - int(h / 2))  # Ajustar a região acima do rosto
#     combined_height = h + (y - hair_y)  # Altura combinada de cabelo e rosto

#     # Desenhar um único retângulo ao redor da região combinada de rosto e cabelo
#     cv2.rectangle(imagem, (x, hair_y), (x + w, hair_y + combined_height), (255, 0, 0), 2)


# Mostrar a imagem com as faces detectadas
from matplotlib import pyplot as plt
plt.imshow(imagem)
plt.axis('off')
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
